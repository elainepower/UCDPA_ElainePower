# Tweet Content Analysis

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd
%matplotlib notebook
```

```python
pancake_mentions = pd.read_csv("PancakeMentions.csv")
pancake_mentions.head()
```

```python
comment_words = '' 
stopwords = set(STOPWORDS)
stopwords = ['nan', 'NaN', 'Nan', 'NAN','QT','RT',"https",'co','slut','t','yup_dilfs'] + list(STOPWORDS)
```

```python
values = pancake_mentions['Hit Sentence'].values

for val in values: 
    val = str(val) 
    tokens = val.split() 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower()
    comment_words += ' '.join(tokens)+' '
```

```python
facecolor = 'white'

wordcloud = WordCloud(width=1000, height=600, 
            background_color=facecolor, 
            stopwords=stopwords,
            min_font_size=15).generate(comment_words)
```

```python
plt.figure(figsize=(10,6), facecolor=facecolor) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.tight_layout(pad=2)
```
