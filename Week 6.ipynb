## GROUPING

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.groupby('sex').mean()
```

```python
# Can't do tips.groupby('sex') because it doesn't know how to aggregate the data
```

```python
tips.groupby(['sex','time']).mean()
```

```python
tips.groupby(['sex','time']).agg(['mean','median'])
```

```python
tips.groupby(['sex','time'])['tip'].agg(['mean','median','std'])
```

```python
tips.groupby(['sex','time']).mean().reset_index()
```

## JOINING

```python
import pandas as pd
```

```python
q1  = pd.DataFrame({"Month": ['Jan', 'Feb', 'Mar'],
                    "Sales": [1200, 1000, 1100]})

q2  = pd.DataFrame({"Month": ['April', 'May', 'Jun'],
                    "Sales": [1300, 1400, 1350]})
```

```python
q1
```

```python
q2
```

```python
pd.concat([q1,q2])
```

```python
pd.concat([q1,q2], ignore_index=True)
```

```python
#Above is to reindex the numbers (ignore_index=True)
```

```python
pd.concat([q1,q2], keys=['Q1','Q2'])
```

```python
#Above is to preserve the individual indexes... so you can split Q1 & Q2
```
